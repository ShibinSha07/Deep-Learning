```python
import tensorflow_hub as hub
import tensorflow_text as text
```

```python
preprocessor_url="https://kaggle.com/models/tensorflow/bert/TensorFlow2/en-uncased-preprocess/3"
encoder_url ="https://www.kaggle.com/models/tensorflow/bert/TensorFlow2/bert-en-uncased-l-10-h-128-a-2/2"
```

```python
bert_preproccess_model=hub.KerasLayer(preprocessor_url)
```

```python
text_test=['nice movie indeed','i love python programming']
text_preprocessed=bert_preproccess_model(text_test)
text_preprocessed.keys()
```

```python
bert_model=hub.KerasLayer(encoder_url)
bert_result=bert_model(text_preprocessed)
bert_result.keys()
```
